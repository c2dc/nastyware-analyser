```python
import sys
import os
import shutil
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dir_1 = '/archive/files/nastyware-files/import-arquivos-nilson'
dir_2 = '/archive/files/nastyware-files/import-malware-bazaar-2021-01-to-2021-02'
dir_3 = '/archive/files/nastyware-files/import-malware-bazaar-2021-01-to-2021-02-f'
dir_4 = '/archive/files/nastyware-files/import-windows-server-2016'
dir_5 = '/archive/files/nastyware-files/import-windows-server-2016-f'
dir_6 = '/archive/files/nastyware-files/import-windows-server-2019'
dir_7 = '/archive/files/nastyware-files/import-windows-server-2019-f'
dir_8 = '/archive/files/nastyware-files/import-windows-server-2022'
dir_9 = '/archive/files/nastyware-files/import-windows-server-2022-f'
dir_10 = './arquivos-nilson/'
dir_11 = './arquivos-nilson-filtro/'
dir_12 = './arquivos-nilson-filtro-2/'
```

```python
dir = dir_9

file_sizes = []
for file in os.listdir(dir):
    file_sizes.append(os.path.getsize(os.path.join(dir, file)))
sns.histplot(file_sizes, kde=True, stat='density', linewidth=0, fill=False)
plt.xlabel('File size (bytes)')
plt.xticks(np.arange(0, 100000, 10000))
plt.ylabel('Density')
plt.title('Distribution of file sizes')
plt.show()
```

```python
dir = dir_9

file_line_counts = []
for file in os.listdir(dir):
    with open(os.path.join(dir, file), 'r') as f:
        file_line_counts.append(len(f.readlines()))

sns.histplot(file_line_counts, kde=True, linewidth=0, fill=False)
plt.xlabel('File line count')
plt.ylabel('Density')
plt.title('Distribution of file sizes')
plt.show()
```

```python
dirs = [dir_9, dir_10, dir_11]

# Plot 3 distributions in one figure using plt.hist
for dir in dirs:
    file_sizes = []
    for file in os.listdir(dir):
        file_sizes.append(os.path.getsize(os.path.join(dir, file)))
    sns.histplot(file_sizes, kde=True, fill=False, linewidth=0, label=dir)

plt.xlabel('File size (bytes)')
plt.ylabel('Density')
plt.title('Distribution of file sizes')
plt.legend()
plt.show()

    
```

```python
dirs = [dir_11, dir_12]

# Plot 3 distributions in one figure using plt.hist
for dir in dirs:
    file_line_counts = []
    for file in os.listdir(dir):
        with open(os.path.join(dir, file), 'r') as f:
            file_line_counts.append(len(f.readlines()))
    sns.histplot(file_line_counts, kde=True, fill=False, linewidth=0, label=dir)

plt.xlabel('Quantidade de linhas')
plt.ylabel('Numero de arquivos')
plt.title('Distribution de quantidade de linhas')
# plt.ylim(0, 800)
plt.legend()
plt.rcParams["figure.figsize"] = (20,5)
plt.show()
```
